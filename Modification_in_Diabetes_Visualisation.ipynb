* **Scatter Plot** :
Individual data points are placed along two axes in a scatter plot graph to show the relationship and distribution of two continuous variables. It visually depicts patterns, correlations, or clusters within data, assisting in understanding the strength of the connection and trend direction between the displayed variables. The spread of dots across the graph depicts the variability and potential relationships between the variables under consideration.


# Scatter Plot

The scatter plot visually represents the relationship between Tri Fold Thick and BMI from the Diabetes Database.

 It illustrates the correlation or dispersion between these variables, aiding in identifying patterns or trends within the dataset for potential insights into their association.

```python
# Scatter plot
sns.scatterplot(x='Tri Fold Thick', y='BMI', data= dg)
plt.title('Scatter Plot of Age vs BMI')
plt.show()

```

* **Histogram**:A histogram is a graphical representation that shows the frequency or count of values within each bin of numerical data by dividing it into intervals called bins. It depicts the data's shape, central tendency, and dispersion visually, providing insights into the data's distribution pattern. Histograms are important for spotting patterns and outliers in a dataset, as well as analyzing the overall distribution properties of the data.


```python
# Histogram
sns.histplot(data=dg, x='PG Concentration', bins=20, color='blue', alpha=0.9, label='PG Concentration')
sns.histplot(data=dg, x='Pregnancies', bins=20, color='red', alpha=0.9, label='Pregnancies')
plt.title('Histogram of PG Concentration and Pregnancies')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.legend()
plt.show()


```

* **Box Plot** : A box plot, also known as a box-and-whisker plot, depicts the distribution of numerical data as quartiles, highlighting the median, interquartile range (IQR), and outliers. It depicts the data's central tendency, spread, and skewness, allowing for fast comparisons between groups or datasets. Box plots provide a clear visual description of the variability of the data, demonstrating dispersion and the presence of extreme values.


```python
# Box plot
sns.boxplot(data=dg[['Age', 'Pregnancies']])
plt.title('Box Plot of Age and feature2')
plt.show()
sns.boxplot(data=dg[['PG Concentration','Diastolic BP']])
plt.title('Box Plot of Age and feature2')
plt.show()
sns.boxplot(data=dg[['Serum Ins','BMI','DP Function']])
plt.title('Box Plot of Age and feature2')
plt.show()
```

## Corelation Matrix

A correlation matrix is a table that displays the pairwise correlations between variables in a dataset, with values ranging from -1 to 1, indicating the strength and direction of their associations. It allows for the detection of relationships between several variables, which aids in understanding how changes in one variable affect changes in others. The matrix provides useful insights into dependencies, assisting in feature selection, assessing multicollinearity, and identifying critical elements influencing data analysis or prediction models.


```python
correlation_matrix = dg.corr()
plt.figure(figsize=(10, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

## Pairplot Graph repesent

A Pairplot graph, created in Python with the seaborn module, depicts pairwise correlations between numerical variables in a dataset using scatterplots and histograms. It displays a grid of graphs in which each variable is compared to each other variable, providing immediate visual insights on correlations, distributions, and probable patterns in the data. Pairplot graphs are effective for preliminary exploratory data analysis, allowing for a thorough knowledge of the relationships and interactions between variables.


```python
selected_features = ['Age', 'PG Concentration', 'BMI', 'class']
sns.pairplot(dg[selected_features], hue='class', markers=['o', 's'])
plt.suptitle('Pairplot Graph', y=1.00, ha='center', fontsize='large')
plt.show()
```

# Bar Graph

```python
X = dg.drop('class', axis=1)
y = dg['class']

rf_classifier = RandomForestClassifier(random_state=42)
rf_classifier.fit(X, y)

feature_importances = pd.Series(rf_classifier.feature_importances_, index=X.columns)
feature_importances.sort_values(ascending=False, inplace=True)

plt.figure(figsize=(10, 8))
feature_importances.plot(kind='bar')
plt.title('Feature Importances')
plt.show()
```

# Pie Graph

```python
X = dg.drop('class', axis=1)
y = dg['class']

rf_classifier = RandomForestClassifier(random_state=42)
rf_classifier.fit(X, y)

feature_importances = pd.Series(rf_classifier.feature_importances_, index=X.columns)
feature_importances.sort_values(ascending=False, inplace=True)

plt.figure(figsize=(10, 8))
feature_importances.plot(kind='pie')
plt.title('Feature Importances')
plt.show()
```

Time Series graphs are critical for forecasting, monitoring, and interpreting temporal activity in many domains, including economics, finance, and weather forecasting.
